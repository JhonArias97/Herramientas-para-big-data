```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Diagramas de dispecion (scatterplot)

from cProfile import label
from turtle import color


x = np.arange(2, 20)
y = x**2
y2 = x**3
plt.scatter(x, y, color="green", marker="*", label="x^2")
plt.scatter(x, y2, color="red", marker="s", label="x^3")
plt.plot(x, y2, color="pink", label="line")


plt.title("Funcion polinomicas", fontsize=20, color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.text(5, 4000, "texto", color="blue")
```

```python
# Histogramas
from turtle import st


mean = 5
std = 3
n = 10000
a = np.random.normal(mean, std, n)

plt.hist(a, bins=20);

plt.title("Histograma de una variable aleatoria", fontsize=20, color="red")
plt.xlabel("x")
plt.ylabel("frecuencia")
plt.grid()
plt.legend()
plt.text(7, 1400, f"mean {mean}, +/- {std}", color="blue")
```

```python
plt.boxplot(a, vert=False);
```

```python
#Subplots
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(15, 5))
ax[0].hist(a);
ax[1].boxplot(a, vert=False);

fig, ay = plt.subplots(ncols=1, nrows=2, figsize=(10, 10), sharex=True)
ay[0].hist(a);
ay[1].boxplot(a, vert=False);
```

```python
# Barplot

clases = ["gatos", "perros", "ratones"]
totales = [10, 30, 20]

plt.barh(clases, totales)
```

```python
data = pd.DataFrame({
    "animales": clases,
    "totales": totales
})
#data.head()

plots = sns.barplot(x="animales", y="totales", data=data)

for bar in plots.patches:
   
    # Using Matplotlib's annotate function and
    # passing the coordinates where the annotation shall be done
    plots.annotate(format(bar.get_height(), '.1f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 5),
                   textcoords='offset points')
```
